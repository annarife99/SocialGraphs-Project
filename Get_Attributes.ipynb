```python
import re
import pandas as pd

songs = pd.read_csv("Song_list.csv")
DataSongs=pd.DataFrame(songs)
songs_list=list(DataSongs.iloc[:,0])
```

```python
Song=[]
Released=[]
Recorded=[]
Genre=[]
Length=[]
Label=[]
Writer=[]
Producer=[]
Album=[]


#PATTERNS to obtain NODE ATTRIBUTES:
pattern='(scope="row")((.*?))([\w.-]+)((.*?))([\w.-]+)'
album='(from the album)((.*?))(</th>)'
pages_path="./raw_songs/"

for a in range(len(songs_list)):
    node=songs_list[a]
    with open(pages_path+node+".txt","r") as f:
        song=f.read()

    if re.search(album, song):
        album_name=re.search(album, song)
        Album.append(album_name.group(2))
    else:
        Album.append('Unknown')
        

    index=[(m.start(0), m.end(0)) for m in re.finditer(pattern, song)]

    for i in range(0,6):
        text=song[index[i][1]:index[i+1][0]]
        print(text)
        abs_text=text[6:-19]
        if i==2:
            Genre.append(abs_text)
        else:
            Genre.append('Unknown')
            
        if i==3:
            Length.append(abs_text)
        else:
            Length.append('Unknown')
            
        if i==4:
            Label.append(abs_text)
        else:
            Label.append('Unknown')
            
        if i==5:
            Writer.append(abs_text)
        else:
            Writer.append('Unknown')


        if re.search(r'<br', abs_text):
            start_index=re.search(r'<br', abs_text).start()
            abs_text=abs_text[0:start_index]

            if i==0:
                Released.append(abs_text)
            else:
                Released.append('Unknown')
            if i==1:
                Recorded.append(abs_text)
            else:
                Recorded.append('Unknown')
```

### Years for each node

```python
song_titles=[]
song_years=[]

def extract_html(url):
    page = urlopen(url)
    html_bytes = page.read()
    html = html_bytes.decode("utf-8")
    return html

from urllib.request import urlopen

pattern1= 'hreflang="en">(.*?)</a>'
pattern2= 'chronological-date">(.*?)</td>'

for i in range(0,9):
    url = 'https://www.thebeatles.com/songs?page='+str(i)
    html=extract_html(url)
    a=re.findall(pattern1,html)
    b=re.findall(pattern2,html)
    
    for el in a:
        el=el.replace('&#039;', '\'')
        song_titles.append(el)
        
    for el2 in b:
        song_years.append(el2)

print(len(song_titles))
print(len(song_years))


data={'song':song_titles,'year':song_years}
df=pd.DataFrame(data)
df
df.to_csv('index=False')
```
