```python
#All Data

Songs = pd.read_csv("AllAtributes.txt") 
DataSongs=pd.DataFrame(Songs)

Albums = pd.read_csv("AllAlbums.txt")
DataAlbums=pd.DataFrame(Albums)

AllSongs = pd.read_csv("Beatles_Dataset.csv")
DataAllSongs=pd.DataFrame(AllSongs)
```
